```python
with open(f"./input/en-frame-elements.tsv") as file:
    df = pd.read_csv(file, sep='\t')
```

```python
temp = [x for x in df['Effect'] if type(x) == str]
effects = []
for x in temp:
    for y in x.split('|'):
        effects.append(y.strip())
        
effects.sort()
len(effects)
```

```python
# removing starting articles/preposition
REG = r'(?i)^((in order )?to|a|that|the|which|will|with|there is|produced?|caused?|of) '

effects = [re.sub(REG, '', x) for x in effects]
effects
```
